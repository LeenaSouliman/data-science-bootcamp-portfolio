```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

```python
print("The average income baseed on Ethnicty is: ")
df[['Ethnicity','Income']].groupby(['Ethnicity']).mean()
```

```python
print("On average single people have higer balance. ")
```

```python
df[['Married','Balance']].groupby(['Married']).mean()
```

```python
print("The highest income in the dataset is 186.634")
df.sort_values(by='Income',ascending=True).tail(1)
```

```python
print("The lowest income in the dataset is 10.354")
df.sort_values(by='Income',ascending=True).head(1)
```

```python
print("There are 1183 cards in the dataset.")
df['Cards'].sum()
```

```python
print("There are 207 number of females we have information on vs 193 number of males.")
df.groupby('Gender').size()
```
