```python
#import packages 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
%matplotlib inline

#load data set, csv and view the colunms to view names and strucutre 
df = pd.read_csv("insurance.csv")
df.head()
```

```python
x = df.iloc[:,0].values.reshape(-1,1) #All rows for colunm index 0, age reshape for 1D array
y = df.iloc[:,6].values #All rows for colunm index 6, i.e charges 
```

```python
# plot data 
plt.scatter(x,y,color = 'g')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.title('Insurance Charges vs Age')
plt.show()
plt.close
```

```python
# fit data to linear regression model and predict output of y, with input of x
# add best-fit line use different colour. 

model = LinearRegression()
model.fit(x,y)
pred = model.predict(x)

plt.scatter(x,y,color = 'g')
plt.plot(x, pred, color = 'r')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.title('Insurance Charges vs Age')
plt.show()
```

```python
# Predict an unknown value
unk_x = [[70]]

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = model.predict(x_pred)

plt.scatter(x,y,color = 'g')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.title('Insurance Charges vs Age')
plt.show()
```
